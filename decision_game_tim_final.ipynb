Description:
    This game is based on my Bachelor Thesis that describes a process of how to manage risks in product development projects/project portfolios. While it is a lot more complex in the real world, this game should show you how it is to manage such risky but also very innovative projects. As it leads you through the different stages you go from defining your idea and evaluating it before starting to work on it (Stage 1), to dealing with upcoming risks during product development (Stage 2), to finally doing some last steps before launching it (Stage 3). Thus, the goal of this game is to bring your project to an end which in this case is defined as a successful launch of the product. Failure on the other hand means that your project was stopped due to economic or risk related reasons.
    
Bugs:
    There are no known game breaking bugs. However, there are some minor issues that are not fixed yet. First, for the project idea it is possible to make no input and still get through this part of stage one (If selecting 'no' for the business idea). Second, in the third stage you can in theory get through the stage by doing the same thing twice(while it is not visible anymore after choosing an option once, you can in theory still type it in again).
